```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['shelf']
collection = db['books']
```

```python
# Добавление одной книги
new_book = {
    "title": "Новая книга",
    "price": 500,
    "description": "Интересное описание"
}

insert_result = collection.insert_one(new_book)
print("Inserted book ID:", insert_result.inserted_id)
```

```python
# Добавление нескольких книг
multiple_books = [
    {"title": "Книга 1", "price": 300, "stock": 10, "description": "No description"},
    {"title": "Книга 2", "price": 300, "stock": 5, "description": "No description"},
    {"title": "Книга 3", "price": 300, "stock": 7, "description": "No description"},
    {"title": "Книга 4", "price": 300, "stock": 4, "description": "No description"}
]
insert_many_result = collection.insert_many(multiple_books)
print("Inserted book IDs:", insert_many_result.inserted_ids)
```

```python
# Книги по заданной цене
eq_books = collection.find({"price": {"$eq": 7.99}})
for book in eq_books:
    print(book['title'])
```

```python
# Книги дороже заданной цены
gt_books = collection.find({"price": {"$gt": 55}})
book = 0
for book in gt_books:
    print(f"{book['title']} - {book['price']}")
```

```python
# Книги дешевле заданной цены
lt_books = collection.find({"price": {"$lt": 15}})
for book in lt_books:
    print(f"{book['title']} - {book['price']}")
```

```python
# Книги с ценой в интервале
gte_lte_books = collection.find({"price": {"$gte": 1, "$lte": 14}})
for book in gte_lte_books:
    print(f"{book['title']} - {book['price']}")
```

```python
# Оператор IN - книги с указанными ценами
in_books = collection.find({"price": {"$in": [25.77, 450]}})
for book in in_books:
    print(f"{book['title']} - {book['price']}")
```

```python
# Оператор NIN - книги, цена которых не равна указанной
nin_books = collection.find({"price": {"$nin": [25, 500]}})
for book in nin_books:
    print(f"{book['title']} - {book['price']}")
```

```python
# Обновление одной книги
update_result = collection.update_one(
    {"title": "Книга 1"},
    {"$set": {"price": 650, "stock": 15}}
)
print("Обновлено книг:", update_result.modified_count)
```

```python
# Обновление множества книг
update_many_result = collection.update_many(
    {"price": {"$eq": 300}},
    {"$inc": {"price": 45}}  # на сколько увеличена цена
)
print("Обновлено книг:", update_many_result.modified_count)
```

```python
# Удаление одной книги
delete_result = collection.delete_one({"title": "Книга 2"})
print("Удалено книг:", delete_result.deleted_count)
```

```python
# Удаление нескольких книг
delete_many_result = collection.delete_many({"price": {"$eq": 345}})
print("Удалено книг:", delete_many_result.deleted_count)
```

```python
# Книги дороже 25 и в наличии менее 5 штук
complex_query = collection.find({
    "$and": [
        {"price": {"$gt": 25}},
        {"stock": {"$eq": 0}}
            ]
})
for query in complex_query:
    print(f"{query['title']}")
```

```python
# Регулярные выражения в поиске
regex_books = collection.find({
    "title": {"$regex": "man"}
})
for regex in regex_books:
    print(f"{regex['title']}")
```

```python
# Сортировка и лимит
sorted_books = collection.find().sort("price", -1).limit(5)
for sorted in sorted_books:
    print(f"{sorted['title']}, - {sorted['price']}")
```
