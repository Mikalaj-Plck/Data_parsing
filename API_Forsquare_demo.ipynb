```python
data = response.json()
places = []

for place in data.get('results', []):
  place_info = {

      'name': place.get('name', 'Название не указано'),
      'address': place.get('location', {}).get('address', 'Адрес не указан'),
      'rating': place.get('rating', 'Рейтинг не доступен')
  }
  places.append(place_info)

category = input("Введите категорию заведений (например, кофейни, музеи, парки): ")

if places:
  print(f"\nНайдены заведения по категории '{category}':")
  for i, place in enumerate(places, 1):
    print(f"\n{i}. {place['name']}")
    print(f"   Адрес: {place['address']}")
    print(f"   Рейтинг: {place['rating']}")
  else:
    print("Заведения не найдены.")
```
